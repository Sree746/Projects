```python
df_modern = df_modern.drop(columns = ["track_number", "track_popularity", "track_duration_min", "explicit", "album_id", "album_release_date", "album_total_tracks", "album_type"])
```

```python
merged_df = base_df.merge(
    df_modern,
    on = "track_id",
    how = "left",
    suffixes = ("_clas", "_mod")
)
```

```python
merged_df.head()
```

```python
cols_to_merge = ["track_name", "artist_name", "artist_popularity", "artist_followers",
                "artist_genres", "album_name"]
for col in cols_to_merge:
    merged_df[f"{col}"] = merged_df[f"{col}_clas"].fillna(
        merged_df[f"{col}_mod"]
    )
    merged_df = merged_df.drop(columns = [f"{col}_clas", f"{col}_mod"])
```

```python
merged_df.head()
```

```python
merged_df.isna().sum()
```

```python
merged_df = merged_df.drop(columns = ["album_total_tracks", "track_number"])
```

```python
merged_df = merged_df.drop_duplicates(subset = ["track_name", "artist_name"]).reset_index(drop = True)
```

```python
merged_df = merged_df.dropna().reset_index(drop = True)
merged_df["release_year"] = pd.to_datetime(
    merged_df["album_release_date"], 
    errors = "coerce"
).dt.year
clean_df = merged_df.drop("album_release_date", axis = 1)
```

```python
clean_df.isna().sum()
```

```python
clean_df["release_year"].describe()
```

```python
median_year = clean_df["release_year"].median()
clean_df["release_year"] = clean_df["release_year"].fillna(median_year)
```

```python
clean_df["explicit"] = clean_df["explicit"].astype(int)
num_features = [
    "track_duration_ms", "track_popularity", "explicit", "artist_popularity", 
    "artist_followers", "release_year"
]
```

```python
text_features = ["artist_genres"]
```

```python
clean_df["artist_genres_clean"] = (clean_df["artist_genres"].str.lower().str.replace(",", " "))
clean_df["artist_genres_clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 1000, stop_words = "english")
```

```python
text_feat_vect = tfidf.fit_transform(clean_df["artist_genres_clean"])
```

```python
text_feat_vect.shape
```

### NORMALIZATION

```python
from sklearn.preprocessing import normalize
```

```python
text_norm = normalize(text_feat_vect)
genre_weight = 0.4
text_norm_weighted = text_norm*genre_weight
```

```python
text_norm_weighted.shape
```

```python
type(text_norm_weighted)
```

### SCALING

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
clean_df["explicit"] = clean_df["explicit"].astype(int)
```

```python
num_scaled = scaler.fit_transform(clean_df[num_features])
```

```python
num_scaled.shape
```

```python
num_feat_weights = np.array([0.4, 2.2, 0.5, 0.5, 0.5, 0.7])
num_weighted = num_scaled * num_feat_weights
```

```python
num_weighted.shape
```

```python
from scipy.sparse import hstack
```

```python
overall_features = hstack((text_norm_weighted, num_weighted)).tocsr()
```

```python
overall_features.shape
```

```python
type(overall_features)
```

### COSINE SIMILARITY

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_songs(song_name, top_n = 10):
    idx = clean_df[clean_df["track_name"].str.lower() == song_name.lower()].index
    if len(idx) == 0:
        return "Song not found"
    idx = idx[0]
    sim_scores = cosine_similarity(
        overall_features[idx], overall_features
    ).flatten()
    top_songs_indices = np.argsort(sim_scores)[::-1][1: top_n]
    return clean_df.loc[top_songs_indices, ["track_name", "artist_name"]]
```

```python
recommend_songs("Shape of you")
```

```python
recommend_songs("Hurt")
```

```python
recommend_songs("Counting Stars")
```

```python
recommend_songs("attention")
```
